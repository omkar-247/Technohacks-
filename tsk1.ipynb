Clean a dataset by removing missing values
and outliers.


Test Table

```python
df = pd.read_csv('test.csv')
df
```

```python
df.isnull().sum()
```

```python
#replacing null values
df['Age'].fillna(df['Age'].median() , inplace= True)
df.dropna(subset = ['Fare'],inplace = True)
df.drop(columns = ['Cabin'],inplace = True)
df.isnull().sum()
```

```python
#finding outliers using box plots
df['PassengerId'].value_counts().plot(kind='box')
plt.title('PassengerId')
plt.show()
df['Pclass'].value_counts().plot(kind='box')
plt.title('Pclass')
plt.show()
df['Name'].value_counts().plot(kind='box')
plt.title('Name')
plt.show()
df['Sex'].value_counts().plot(kind='box')
plt.title('Sex')
plt.show()
df['Age'].value_counts().plot(kind='box')
plt.title('Age')
plt.show()
df['SibSp'].value_counts().plot(kind='box')
plt.title('SibSp')
plt.show()
df['Parch'].value_counts().plot(kind='box') 
plt.title('Parch') 
plt.show()
df['Ticket'].value_counts().plot(kind='box')
plt.title('Ticket')
plt.show()
df['Fare'].value_counts().plot(kind='box')
plt.title('Fare')
plt.show()
df['Embarked'].value_counts().plot(kind='box')
plt.title('Embarked')
plt.show()
```

```python
#removing outliers using IQR method
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3-Q1
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
df[(df['Fare']>= lower_bound) & (df['Fare']<= upper_bound)]
```
