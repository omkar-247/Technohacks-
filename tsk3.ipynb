Task 3 : Visualization using Histogram
Create a histogram or bar chart to visualize
the distribution of data in a dataset

```python
plt.hist(df['SepalLengthCm'],alpha = 0.8,edgecolor = 'k')
plt.xlabel('Sepal Length in Cm')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['SepalWidthCm'],alpha = 0.8,edgecolor = 'k')
plt.xlabel('Sepal Width in Cm')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['PetalLengthCm'],alpha = 0.8,edgecolor = 'k')
plt.xlabel('Sepal Length in Cm')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['PetalWidthCm'],alpha = 0.8,edgecolor = 'k')
plt.xlabel('Sepal Width in Cm')
plt.ylabel('Frequency')
plt.show()
```

```python
species_count = df['Species'].value_counts()
plt.bar(species_count.index,species_count.values ,color = ['red','blue','green'])
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.title('Distribution of Species')
```
